```python
class Node:
    def __init__(self,data):
        self.data = data
        self.link  = None 
class Queue:
    def __init__(self,front,rear):
        self.head = None
        self.front = front 
        self.rear = rear
    def enqueue(self,data):
        new = Node(data)
        if self.head is None:
            self .head = new
            self.front = self.front + 1
            self.rear = self.rear + 1
            print("Front:{0} Rear:{1}".format(self.front,self.rear))
        else:
            temp = self.head
            while temp.link is not None:
                temp = temp.link 
            temp.link = new 
            self.rear = self.rear + 1
            print("Front:{0} Rear:{1}".format(self.front,self.rear))
    def dequeue(self):
        print("Dequeued element:",self.head.data)
        self.head = self.head.link 
        self.rear = self.rear - 1
        print("Front:{0} Rear:{1}".format(self.front,self.rear))
    def check(self):
        temp = self.head
        print("First element:",self.head.data)
        while temp.link is not None:
            temp = temp.link 
        print("Last element:",temp.data)
        print("Front:{0} Rear:{1}".format(self.front,self.rear))
    def display(self):
        print("Display:")
        temp = self.head
        while temp is not None :
            print(temp.data)
            temp = temp.link
maxsize = int(input("Enter size:"))
front = 0
rear = 0
obj  = Queue(front,rear)

for i in range(0,maxsize):
    if i < maxsize:
        x=int(input("Enter data : "))
        obj.enqueue(x)
        i = i + 1 
obj.display()
obj.dequeue()
obj.display()
obj.check()
obj.dequeue()
obj.display()
obj.check()
```
